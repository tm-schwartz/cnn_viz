```python
def heart_data():
    df = pd.read_csv("https://raw.githubusercontent.com/grbruns/cst383/master/heart.csv")
    predictors = ['age', 'restbp', 'chol', 'maxhr']
    X = df[predictors].values
    y = (df['output'] - 1).values
    return X,y,'heart'
```

```python
class CheckpointCallback(tf.keras.callbacks.Callback):
	def __init__(self):
		super(CheckpointCallback, self).__init__()
		self.epoch = 0

	def on_epoch_end(self, batch, logs=None):
		self.model.save_weights(f'INTERPOLATIONDATACHECK/model_{self.epoch}_heartdata.h5')
		self.epoch += 1
```

```python
def build_model(X, y, n_hidden, learning_rate=0.01):
    """ Build a feedforward network with one hidden layer """
    input_shape = X.shape[1:]
    n_outputs = np.unique(y).size 
    inputs  = tf.keras.layers.Input(shape=input_shape)
    dense1   = tf.keras.layers.Dense(n_hidden//2, activation='selu')(inputs)
    dense2   = tf.keras.layers.Dense(n_hidden, activation='tanh')(dense1)
    dense3   = tf.keras.layers.Dense(n_hidden*2, activation='tanh')(dense2)
    outputs = tf.keras.layers.Dense(n_outputs, activation='softmax', name='output')(dense3)
    model   = tf.keras.models.Model(inputs=inputs, outputs=outputs)
    model.compile(loss='sparse_categorical_crossentropy', 
                optimizer=tf.keras.optimizers.SGD(lr=learning_rate), 
                metrics=['accuracy'])
    return model
```

```python
import os
def sparse_categorical_crossentropy(target, output):
    epsilon = 1e-7
    output = clip(output[:,1], epsilon, 1-epsilon)
    bce = target * np.log(output) + (1 - target) * np.log(1 - output)
    return np.mean(-bce)

def clip(x, lo, hi):
    return np.maximum(np.minimum(x, hi), lo)
# compute loss along interpolated line from start to end
# (following Goodfellow paper we actually go past the end)
def _build_model(X, y, n_hidden, w, learning_rate=0.01):
    """ Build a feedforward network with one hidden layer """
    input_shape = X.shape[1:]
    n_outputs = np.unique(y).size 
    inputs  = tf.keras.layers.Input(shape=input_shape, dtype='float64')
    dense1   = tf.keras.layers.Dense(n_hidden//2, activation='selu',dtype='float64')(inputs)
    dense2   = tf.keras.layers.Dense(n_hidden, activation='tanh', dtype='float64')(dense1)
    dense3   = tf.keras.layers.Dense(n_hidden*2, activation='tanh', dtype='float64')(dense2)
    outputs = tf.keras.layers.Dense(n_outputs, activation='softmax', name='output', dtype='float64')(dense3)
    model   = tf.keras.models.Model(inputs=inputs, outputs=outputs)
    model.compile(loss='sparse_categorical_crossentropy', 
                optimizer=tf.keras.optimizers.SGD(lr=learning_rate), 
                metrics=['accuracy'])
    model.load_weights(w)
    return model

weights = [m.get_weights() for m in [_build_model(X, y, 100, f'INTERPOLATIONDATACHECK/{w}') for w in sorted(os.listdir('INTERPOLATIONDATACHECK/'), key=lambda k: int(k.split('_')[1]))]]
n = 100
loss_train_int = np.zeros(n)
loss_test_int = np.zeros(n)
alphas = np.zeros(n)
model = _build_model(X, y, 100, f'INTERPOLATIONDATACHECK/model_0_heartdata.h5')
for i,alpha in enumerate(np.linspace(0.0, 1.0, 100)):
    alphas[i] = alpha
    w = interpolate_check(weights[0], weights[-1], alpha)
    model.set_weights(w)
    y_probs = model.predict(X_train)
    loss_train_int[i] = sparse_categorical_crossentropy(y_train, y_probs)
    y_probs = model.predict(X_test)
    loss_test_int[i] = sparse_categorical_crossentropy(y_test, y_probs)

# compute loss along training path
m = len(weights)
loss_train = np.zeros(m)
loss_test = np.zeros(m)
for i,w in enumerate(weights):
    model.set_weights(w)
    y_probs = model.predict(X_train)
    loss_train[i] = sparse_categorical_crossentropy(y_train, y_probs)
    y_probs = model.predict(X_test)
    loss_test[i] = sparse_categorical_crossentropy(y_test, y_probs)
```
